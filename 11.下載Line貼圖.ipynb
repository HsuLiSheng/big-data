```python
import os
import json
import requests
from bs4 import BeautifulSoup
```

```python
img_dir = 'line stickers/NL'
if not os.path.exists(img_dir):
  os.mkdir(img_dir)
```

```python
url = 'https://store.line.me/stickershop/product/13117151/zh-Hant'
html = requests.get(url)
soup = BeautifulSoup(html.text, 'html.parser')

datas = soup.find_all('li', {'class':'mdCMN09Li FnStickerPreviewItem'})

print(type(datas[0]))
print(datas[0])
```

```python
for data in datas:
  ###data-preview的屬性內容為字典格式的字串，要用json讀取把字串換成字典
  #可以用.get或是[]方式拿到data-preview的屬性內容
  #infor = json.loads(data.get('data-preview'))
  infor = json.loads(data['data-preview'])


  id = infor['id']
  ###載入圖片
  img_file = requests.get(infor['staticUrl'])

  path = os.path.join(img_dir,id)
  ###wb:以二進制的方式打開文件寫內容進去
  with open(path+'.png', 'wb') as f:
    f.write(img_file.content)
    
  print(path+'.png' + '下載完成')  
```
