```python
import matplotlib.pyplot as plt

list1 = [1,3,5,7,9]
list2 = [15,30,50,20,80]
plt.plot(list1, list2, color='red', lw='5.0', ls=':', label='food') ###lw:線條寬度，ls:線條形式
plt.legend() ###顯示label屬性
plt.show()

list3 = [2,4,6,8,10]
list4 = [35,15,40,85,95]

plt.plot(list1, list2, label='Male')
plt.plot(list3, list4, color='red', lw=5, ls='--', label='Female')
plt.legend()
plt.xlim(0,10)
plt.ylim(0,100)
plt.title('Pocket money')
plt.xlabel('Age')
plt.ylabel('Money')
plt.show()
```

```python
###繪製柱狀圖和圓餅圖

plt.bar(list1, list2, label='Male')
plt.bar(list3, list4, color='red', lw=5, ls='--', label='Female')
plt.legend()
plt.xlim(0,12)
plt.ylim(0,100)
plt.title('Pocket money')
plt.xlabel('Age')
plt.ylabel('Money')
plt.show()

labels = ['South', 'East', 'North', 'West']
sizes = [5,15,10,20]
colors = ['blue', 'red', 'green', 'yellow']
explode = (0,0,0.1,0)
plt.pie(sizes, explode=explode, labels=labels, colors=colors,\
        labeldistance=1.1, autopct='%3.1f%%', shadow=True,\
        startangle=90, pctdistance=0.6)   
###explode:凸出串列(0為正常)
###labeldistance:標題和圓心的距離為半徑的幾倍，autopct:百分比格式
###startangle:繪圖的起始角度, pctdistance:百分比文字和圓心距離為半徑的幾倍

plt.axis('equal') ###繪製正圓形圓餅圖 (預設為橢圓形)
plt.legend()
plt.show()
```

```python
import pandas as pd

data = [[65,92,78,83,70], [90,72,76,93,56], [81,85,91,89,77], [79,53,47,94,80]]
indexs = ['A', 'B', 'C', 'D']
columns = ['Chinese', 'Math', 'English', 'Nature', 'Society']
df = pd.DataFrame(data, columns=columns, index=indexs)
df.plot(kind='bar', title='Test score', fontsize=16) ###kind:繪圖模式，fontsize:xy軸刻度的字體大小
```

```python
import plotly
from plotly.graph_objs import Scatter, Layout

plotly.offline.init_notebook_mode(connected=True) ##讓離線繪圖在jupyter中顯示

data = [Scatter(x=['A', 'B', 'C', 'D'], y=[65,92,78,83],  ###繪圖資料來源
      text=['one', 'two', 'three', 'four'],     ###在繪圖點加入文字
      textposition='top center',
      mode = 'markers+lines+text')]   ###展現顯示屬性
      
plotly.offline.iplot({'data':data, 
           'layout':Layout(title='Test score')}) ###加入標題
```

```python
data = [Scatter(x=['A', 'B', 'C', 'D'], y=[65,92,78,83], name='Chinese'),  ###為每個圖形命名
     Scatter(x=['A', 'B', 'C', 'D'], y=[85,65,70,85], name='Math'),
     Scatter(x=['A', 'B', 'C', 'D'], y=[90,85,95,86], name='English')]  

plotly.offline.iplot({'data':data, 
           'layout':Layout(title='Test score')}) ###加入標題
```
