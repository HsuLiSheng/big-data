```python
import requests
import json
```

```python
url = 'https://trends.google.com/trends/api/dailytrends'
payload = { 'hl': 'zh-TW',
        'tz': '-480',
        ###指定日期，若沒有則取近兩天
        # 'ed': '20221201'
        'geo': 'TW',
        'ns': '15'
      }
html = requests.get(url, params=payload)
html.encoding = 'utf-8'

###數字1代表分割成2份，第一段為多餘文字，以_代替，第二段為內容，存到datas變數中
_,datas = html.text.split(',',1)
###下載的資料轉換為字典
data = json.loads(datas)
```

```python
trendingSearchesDays = data['default']['trendingSearchesDays']
###1.找到近兩天的日期
###2.找到不同的主題關鍵字
###3.找到關鍵字的相關文章

for trendingSearchesDay in trendingSearchesDays:
   date = trendingSearchesDay['formattedDate']
   print('日期:' + date)
   print()

   for item in trendingSearchesDay['trendingSearches']:
     print('主題關鍵字:' + item['title']['query'])
     print()
     for content in item['articles']:
       print('標題:' + content['title'])
       print('媒體:' + content['source'])
       print('發布時間:' + content['timeAgo'])
       print('內容:' + content['snippet'])
       print('相關連結:' + content['url'])
       print()

     print('......................')
```

```python
###1.找到近兩天的日期
###2.找到不同的主題關鍵字
###3.找到關鍵字的相關文章

trendingSearchesDays = data['default']['trendingSearchesDays']
for trendingSearchesDay in trendingSearchesDays:
   news=''
   date = trendingSearchesDay['formattedDate']
   
   news += '日期:' + date + '\n\n'
   
   i=0
   ###蒐集到的全部內容
   for item in trendingSearchesDay['trendingSearches']:
     i += 1
     news += '...........................................................' + '\n'
     news += str(i) + '. '+ '主題關鍵字:' + item['title']['query'] + '\n'
     news += '...........................................................' + '\n\n'

     for content in item['articles']:
       news += '標題:' + content['title'] + '\n'
       news += '媒體:' + content['source'] + '\n'
       news += '發布時間:' + content['timeAgo'] + '\n'
       news += '內容:' + content['snippet'] + '\n'
       news += '相關連結:' + content['url'] + '\n\n'
       
   filename = trendingSearchesDay['date'] + '.txt'
   with open('dailytrends/'+filename, 'w', encoding='utf-8') as f:
     f.write(news)
   print(filename+ '已存檔!!!')    
```
